### Required Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
dataset=pd.read_csv('WineQT.csv')
```

```python
dataset=dataset.drop(['Id'],axis=1)
```

```python
sns.catplot(x='quality',data=dataset,kind='count')
```

```python
# volatile acidity vs quality
plot = plt.figure(figsize=(10,5))
sns.barplot(x='quality',y='volatile acidity',data = dataset)
```

```python
plot = plt.figure(figsize=(10,5))
sns.barplot(x='quality',y='citric acid',data = dataset)
```

```python
# create a heatmap to understand the corr relation between coloums
plt.figure(figsize=(15,8))
sns.heatmap(dataset.corr(),annot=True , cmap="YlGnBu")
```

```python
# Data Preprocessing
```

```python
x=dataset.drop(['quality'],axis=1)
```

```python
x
```

```python
# Label Binarization
```

```python
y=dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0 )
```

```python
y
```

### Split the data into two part

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

### Train the Data using RandomForestRegressor

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
newModel=RandomForestClassifier()
```

```python
newModel.fit(x_train,y_train)
```

```python
newModel.score(x_test,y_test)
```
